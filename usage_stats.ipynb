```python
conda_stats_dfs = []

for stat in conda_stats:
    new_stat = {}
    new_stat['df'] = pd.DataFrame.from_dict({k: v for k, v in stat.items() if k in ['2.7', '3.3', '3.4', '3.5']})
    new_stat['date'] = stat['date']
    new_stat['release'] = stat['release']
    conda_stats_dfs.append(new_stat)
```

```python
conda_stats_dfs
```

```python
conda_major_stats_dfs = {}

for stat in conda_stats_dfs:
    release = stat['release'][0:3]
    #old_df = conda_major_stats_dfs[release]['df']
    try:
        conda_major_stats_dfs[release]['df'] = conda_major_stats_dfs[release]['df'] + stat['df']
    except KeyError:
        conda_major_stats_dfs[release] = {}
        conda_major_stats_dfs[release]['df'] = stat['df']
```

```python
conda_major_stats_dfs
```

```python
ymax = 250

for release, stat in sorted(conda_major_stats_dfs.items()):
    fig, ax = plt.subplots()
    stat['df'].plot(ax=ax, kind='bar', rot=0)
    ax.set_title('PVLIB Python {}'.format(release))
    ax.set_ylim(0, ymax)
    ax.set_ylabel('anaconda.org downloads')
    
for release, stat in sorted(conda_major_stats_dfs.items()):
    fig, ax = plt.subplots()
    stat['df'].sum(axis=1).plot(ax=ax, kind='bar', rot=0)
    ax.set_title('PVLIB Python {}'.format(release))
    ax.set_ylim(0, 250)
    ax.set_ylabel('anaconda.org downloads')
    
for release, stat in sorted(conda_major_stats_dfs.items()):
    fig, ax = plt.subplots()
    stat['df'].sum(axis=0).plot(ax=ax, kind='bar', rot=0)
    ax.set_title('PVLIB Python {}'.format(release))
    ax.set_ylim(0, ymax)
    ax.set_ylabel('anaconda.org downloads')
```
